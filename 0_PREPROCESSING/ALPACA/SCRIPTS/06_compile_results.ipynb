# Join GLOSA + Não Convertidos

```python
df_not_translated = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DATASETS/ALPACA_FILTERED/result_not_to_figurative_joined.csv")
df_not_translated.shape
```

```python
df_not_translated[df_not_translated['ID'] == 6]
```

```python
df_combined = pd.merge(df_glosa, df_not_translated, on='ID', how='outer', suffixes=('_df1', '_df2'))

df_combined['INSTRUCTION'] = df_combined['INSTRUCTION_df1'].combine_first(df_combined['INSTRUCTION_df2'])
df_combined['INPUT'] = df_combined['INPUT_df1'].combine_first(df_combined['INPUT_df2'])
df_combined['OUTPUT'] = df_combined['OUTPUT_df1'].combine_first(df_combined['OUTPUT_df2'])

df_combined = df_combined[['ID', 'INSTRUCTION', 'INPUT', 'OUTPUT']]
df_combined[df_combined['ID'] == 6]
```
