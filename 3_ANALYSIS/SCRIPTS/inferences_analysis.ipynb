# Benchmark

```python
def plot_por_familia_dataset(df, familia_dataset):
    # Sua ordem fixa de modelos
    ordem_modelos = [
        'Llama-3.1-8B', 'Llama-3.1-8B-V1', 'Llama-3.1-8B-V2',
        'Llama-3.2-3B', 'Llama-3.2-3B-V1', 'Llama-3.2-3B-V2',
        'Phi-4-14B', 'Phi-4-14B-V1', 'Phi-4-14B-V2',
        'Qwen2.5-7B', 'Qwen2.5-7B-V1', 'Qwen2.5-7B-V2',
        'Qwen2.5-14B', 'Qwen2.5-14B-V1', 'Qwen2.5-14B-V2',
        'Zephyr-7B', 'Zephyr-7B-V1', 'Zephyr-7B-V2'
    ]

    modelos_destaque = ["Llama-3.1-8B-V1", "Phi-4-14B-V1", "Qwen2.5-14B-V1"]

    # Filtrar apenas os dados do dataset/família desejada
    df_familia = df[df["dataset"] == familia_dataset]

    # Agrupar por modelo e dataset
    df_grouped = df_familia.groupby(["dataset", "modelos"], as_index=False)["pct_acertos"].mean().fillna(0)

    # Criar coluna formatada para destaque
    df_grouped["modelos_formatado"] = df_grouped["modelos"].apply(
        lambda x: f"<b>{x}</b>" if x in modelos_destaque else x
    )

    df_grouped["text_formatado"] = df_grouped.apply(
        lambda row: f"<b>{row['pct_acertos']:.1f}%</b>" if row["modelos"] in modelos_destaque else f"{row['pct_acertos']:.1f}%",
        axis=1
    )

    # A ordem base continua sendo a lista original
    modelos_formatados_na_ordem = []
    for modelo in ordem_modelos:
        if modelo in modelos_destaque:
            modelos_formatados_na_ordem.append(f"<b>{modelo}</b>")
        else:
            modelos_formatados_na_ordem.append(modelo)

    # Converter coluna para categoria ordenada
    df_grouped["modelos_formatado"] = pd.Categorical(
        df_grouped["modelos_formatado"],
        categories=modelos_formatados_na_ordem,
        ordered=True
    )

    # Reordenar o DataFrame com base na ordem definida
    df_grouped = df_grouped.sort_values("modelos_formatado")

    # Criar gráfico de barras horizontais
    fig = px.bar(
    df_grouped,
    y="modelos_formatado",
    x="pct_acertos",
    color="modelos",  # mudar aqui para a coluna original
    color_discrete_map=cores_modelos,
    orientation='h',
    text="text_formatado",
    title=f"Porcentagem Média de Acertos para Dataset '{familia_dataset}' por Modelo"
)


    fig.update_layout(
        width=1200,
        height=700,
        margin=dict(l=150),
        yaxis=dict(
            categoryorder='array',
            # INVERTE A LISTA AQUI 👇
            categoryarray=list(reversed(modelos_formatados_na_ordem)),
            title_font=dict(size=16, color='black'),
            tickfont=dict(size=16, color='black')
        ),
        xaxis=dict(
            title_font=dict(size=16, color='black'),
            tickfont=dict(size=16, color='black')
        ),
        title=dict(
            font=dict(size=16, color='black')
        ),
        plot_bgcolor='white',
        paper_bgcolor='white',
        font=dict(size=16, color='black')
    )

    # Aplica formatação só nas barras
    fig.update_traces(
        selector=dict(type="bar"),
        textposition='outside',
        textfont=dict(size=16, color='black'),
        width=0.6,
        showlegend=False
    )

    fig.show()
    
plot_por_familia_dataset(df_result_final, "Inglês")
```

# Com Fine tune: V1 e V2

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# === CONFIG ===
ordem_datasets = ["Português", "GLOSA e Português", "GLOSA"]
cores_datasets = {
    "Português": "#2ca02c",
    "GLOSA e Português": "#ff7f0e",
    "GLOSA": "#d62728"
}
familias = [
    "Llama-3.1",
    "Llama-3.2",
    "Qwen2.5-7B",
    "Qwen2.5-14B",
    "Zephyr",
    "Phi"
]

# === FUNÇÃO PARA TRANSFORMAR EM NEGRITO (UNICODE) ===
def bold_unicode(text):
    bold_nums = str.maketrans("0123456789.", "𝟬𝟭𝟮𝟯𝟰𝟱𝟲𝟳𝟴𝟵․")
    return text.translate(bold_nums)

# === SUBPLOTS ===
fig = make_subplots(
    rows=3, cols=2,
    subplot_titles=familias,
    shared_yaxes=True,
    vertical_spacing=0.1,
    horizontal_spacing=0.1
)

# Para controlar quais legendas já apareceram
legendas_adicionadas = set()

# === LOOP POR FAMÍLIA ===
for idx, familia in enumerate(familias):
    row = idx // 2 + 1
    col = idx % 2 + 1

    df_fam = df_result_final[
        (df_result_final["modelos"].str.startswith(familia)) &
        (df_result_final["dataset"].isin(ordem_datasets))
    ].copy()

    if df_fam.empty:
        continue

    df_grouped = df_fam.groupby(["modelos", "dataset"], as_index=False)["pct_acertos"].mean().fillna(0)
    modelos_ordenados = sorted(df_grouped["modelos"].unique())

    for modelo in modelos_ordenados:
        df_modelo = df_grouped[df_grouped["modelos"] == modelo]

        destaque = "-V1" in modelo or "-V2" in modelo
        valor_max = df_modelo["pct_acertos"].max() if destaque else None

        for dataset in ordem_datasets:
            df_ds = df_modelo[df_modelo["dataset"] == dataset]
            if df_ds.empty:
                continue

            valor = df_ds.iloc[0]["pct_acertos"]
            texto = f"{valor:.1f}%"
            if destaque and valor == valor_max:
                texto = bold_unicode(texto)

            # Só mostra legenda se for a 1ª vez que este dataset aparece
            showlegend = False if dataset in legendas_adicionadas else True
            legendas_adicionadas.add(dataset)

            fig.add_trace(
                go.Bar(
                    x=[modelo],
                    y=[valor],
                    name=dataset,
                    showlegend=showlegend,
                    marker_color=cores_datasets[dataset],
                    text=[texto],
                    textposition="outside",
                    textfont=dict(size=16),  # correto!
                    offsetgroup=dataset
                ),
                row=row, col=col
            )

    fig.update_xaxes(
        categoryorder="array",
        categoryarray=modelos_ordenados,
        tickangle=0,
        row=row, col=col
    )

# === LAYOUT FINAL ===
fig.update_layout(
    barmode="group",
    height=1300,
    width=1000,
    plot_bgcolor="white",
    paper_bgcolor="white",
    font=dict(color="black", size=16),
    legend=dict(
        orientation="h",
        yanchor="top",
        y=-0.05,
        xanchor="left",
        x=0,
        font=dict(size=16)
    ),
    title_text="Comparação Base vs V1/V2 por Dataset (Maior valor em negrito para V1/V2)",
    title_font=dict(size=16)
)

fig.show()
```

```python
df = df_result_final[df_result_final["dataset"].isin(["GLOSA e Português", "Português"])][["modelos", "dataset", "tipo", "pct_acertos"]]

def comparar_com_baseline(df, modelo_base):
    """
    Compara as versões V1 e V2 (dataset GLOSA e Português)
    com o baseline (V0 do mesmo modelo, dataset GLOSA e Português)
    """
    resultados = []

    # 1) Valor baseline - agora no GLOSA e Português
    baseline_val = df[
        (df["modelos"] == modelo_base) &
        (df["dataset"] == "GLOSA e Português") &
        (df["tipo"] == "sem_fine_tuning")
    ]["pct_acertos"].values

    if len(baseline_val) == 0:
        print(f"Baseline não encontrado para {modelo_base}")
        return pd.DataFrame()

    baseline = baseline_val[0]

    # 2) Versões V1 e V2 em GLOSA+PT
    for versao in ["V1", "V2"]:
        modelo_v = f"{modelo_base}-" + versao

        val = df[
            (df["modelos"] == modelo_v) &
            (df["dataset"] == "GLOSA e Português")
        ]["pct_acertos"].values

        if len(val) == 0:
            pct = None
            ganho = None
        else:
            pct = val[0]
            ganho = pct - baseline

        resultados.append({
            "Modelo": modelo_v,
            "Baseline_GLOSA_PT": baseline,
            "GLOSA_PT": pct,
            "Ganho": ganho
        })

    return pd.DataFrame(resultados)

# ✅ Exemplos
print(comparar_com_baseline(df, "Llama-3.1-8B"))
print(comparar_com_baseline(df, "Llama-3.2-3B"))
print(comparar_com_baseline(df, "Qwen2.5-7B"))
print(comparar_com_baseline(df, "Qwen2.5-14B"))
print(comparar_com_baseline(df, "Phi-4-14B"))
print(comparar_com_baseline(df, "Zephyr-7B"))

```
